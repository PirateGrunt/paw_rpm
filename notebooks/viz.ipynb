## Scatterplot

```python
from matplotlib import pyplot as plt

accident_years = [
      1991
    , 1992
    , 1993
    , 1994
    , 1995
]

ibnr = [
      1100
    , 1200
    , 1300
    , 1400
    , 1500
]
```

```python
plt.plot(
      accident_years
    , ibnr
)
```

Change some visual characteristics with function arguments
Decorate the plot with functions like `title()` and `ylabel()`

```python
plt.plot(
      accident_years
    , ibnr
    , color = 'red'
    , marker = 'o'
    , linestyle = 'dashed'
)
plt.title('IBNR by AY')
plt.xlabel('Accident Year')
plt.ylabel("IBNR ($000s)")
```

## Seaborn

* Advanced alternative to matplotlib
* Smells a little like ggplot2

```python
import seaborn as sns

sns.set()
tips = sns.load_dataset("tips")
sns.relplot(
      x = "total_bill"
    , y = "tip"
    , hue = "smoker"
    , style = "smoker"
    , size = "size"
    , data = tips);
```
