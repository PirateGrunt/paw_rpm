## Scatterplot

```python
from matplotlib import pyplot as plt
# tell Jupyter to plot graphs
%matplotlib inline

accident_years = [
      1991
    , 1992
    , 1993
    , 1994
    , 1995
]

ibnr = [
      1100
    , 1200
    , 1300
    , 1400
    , 1500
]
```

```python
plt.plot(
      accident_years
    , ibnr
)
```

Change some visual characteristics with function arguments
Decorate the plot with functions like `title()` and `ylabel()`

```python
plt.plot(
      accident_years
    , ibnr
    , color = 'red'
    , marker = 'o'
    , linestyle = 'dashed'
)
plt.title('IBNR by AY')
plt.xlabel('Accident Year')
plt.ylabel("IBNR ($000s)")
```

```python
# shorthand for color (r), marker (x) and line style (-- for dashed)
plt.plot(
      accident_years
    , ibnr
    , 'rx--'
)
# or set multiple properties of current axis, gca per Matlab:
plt.gca().set(title='IBNR by AY', xlabel='Accident Year', ylabel="IBNR ($000s)", 
              xlim=[1991, 1995], ylim=[1000, 2000])
```

```python
# if you like TeX
import numpy as np
xs = np.linspace(1/10000, .5, 10000)
ys = np.sin(1/xs)
plt.plot(xs, ys, linewidth=0.5, label='For legend')
plt.gca().set(title='Graph of $\sin(x^{-1})$)')
plt.legend(title='Legend Title')
```

## Seaborn

* Wrapper around matplotlib to make life easier
* Smells a little like ggplot2

> If matplotlib tries to make easy things easy and hard things possible, seaborn tries to make a well-defined set of hard things easy too. 
  (Michael Waskom, *Introduction to Seaborn*)

```python
import seaborn as sns
# need searborn 0.9.0 or later for relplot

sns.set()
tips = sns.load_dataset("tips")
sns.relplot(
      x = "total_bill"
    , y = "tip"
    , hue = "smoker"
    , style = "smoker"
    , size = "size"
    , data = tips);
```

```python
for style in ['dark', 'white', 'darkgrid', 'whitegrid']:
    sns.set(style=style)
    tips = sns.load_dataset("tips")
    sns.relplot(
          x = "total_bill"
        , y = "tip"
        , hue = "smoker"
        , style = "smoker"
        , size = "size"
        , data = tips)
    plt.title(style);
```
